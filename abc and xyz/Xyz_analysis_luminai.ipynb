```python
# Group by Material and sort within each group by Material_Id
sorted_df = df_wholedata.sort_values(['Material'])

# Display sorted DataFrame
display(sorted_df)
```

```python
### INTERMITTENT PRODUCTS ARE REMOVED (PRODUCTS WHO DO NOT HAVE A CONSISTENT DATA FOR THE TIME PERIOD 25TH MONTH TO 36TH MONTH)


# Create a set of all time IDs from 25 to 36
required_time_ids = set(range(25, 37))

# Group the dataframe by Material and check if all required_time_ids are present for each Material
valid_materials = df_wholedata.groupby('Material')['Time id'].apply(lambda x: set(x) >= required_time_ids)

# Filter out records based on valid materials
filtered_df = sorted_df[sorted_df['Material'].isin(valid_materials.index[valid_materials])]

# Display filtered DataFrame
display(filtered_df)

```

```python
# Move the 'Quantity' column to the last position
filtered_df = filtered_df[[col for col in filtered_df.columns if col != 'Quantity'] + ['Quantity']]

# Display DataFrame with 'Quantity' column moved to the last position
display(filtered_df)
```

```python
# Copy the "Quantity" column to create the "lags" column
filtered_df['lags'] = filtered_df['Quantity'].shift(1)

# Display the DataFrame with the new "lags" column
display(filtered_df)
```

```python
# Calculate percentage difference and store the result in a new column named "percentage_diff"
filtered_df['percentage_diff'] = ((filtered_df['Quantity'] - filtered_df['lags']) / filtered_df['lags']) * 100

# Display the DataFrame with the new "percentage_diff" column
(filtered_df)
```

```python
# Calculate average percentage difference for each unique Material
average_percentage_diff = filtered_df.groupby(['Material', 'Material_Id'])['percentage_diff'].mean().reset_index()

# Rename the column
average_percentage_diff.rename(columns={'percentage_diff': 'average_percentage_diff'}, inplace=True)

# Display the new dataframe
display(average_percentage_diff)
```

```python
# Define the new bins and labels for xyz_analysis
bins = [-float('inf'), 35, 60, float('inf')]
labels = ['x', 'y', 'z']

# Create the xyz_analysis column based on the new bins and labels
average_percentage_diff['xyz_analysis'] = pd.cut(average_percentage_diff['average_percentage_diff'], bins=bins, labels=labels, include_lowest=True)

# Display the updated dataframe with the new xyz_analysis column
display(average_percentage_diff)


# Count how many materials are categorized into each 'x', 'y', and 'z' category
xyz_counts = average_percentage_diff['xyz_analysis'].value_counts()

# Display the counts
print(xyz_counts)

```

```python
# Create separate dataframes for each category in xyz_analysis
df_x = average_percentage_diff[average_percentage_diff['xyz_analysis'] == 'x']
df_y = average_percentage_diff[average_percentage_diff['xyz_analysis'] == 'y']
df_z = average_percentage_diff[average_percentage_diff['xyz_analysis'] == 'z']

# Display the dataframes
print("Dataframe for category 'x':")
display(df_x)

print("\nDataframe for category 'y':")
display(df_y)

print("\nDataframe for category 'z':")
display(df_z)
```
