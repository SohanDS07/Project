```python
import pandas as pd
```

```python
df= pd.read_csv('ax_df_all.csv')
df
```

# Modelling
#Naive Forecasting

```python
# Apply Naive forecasting within each group
df['Naive Forecast_qty'] = df.groupby('Material')['Quantity'].shift(1)
```

```python
df
```

# Simple Average for 3 month and 6 months


```python
# Apply Simple Average forecasting within each group for 3-month and 6-month periods
df['Simple Average 3-month_qty'] = df.groupby('Material')['Quantity'].rolling(window=3).mean().reset_index(0, drop=True)
df['Simple Average 6-month_qty'] = df.groupby('Material')['Quantity'].rolling(window=6).mean().reset_index(0, drop=True)
```

```python
df
```

```python
# Define the window size for the moving average
window_size = 3  # You can adjust this as needed

# Apply Moving Average forecasting within each group
df['Moving Average'] = df.groupby('Material')['Quantity'].rolling(window=window_size).mean().reset_index(0, drop=True)

```
